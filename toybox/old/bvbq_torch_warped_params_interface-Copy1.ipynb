```python
import sys
sys.path.append('..')
import functools
import math
import collections

import numpy as np
import torch
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import sobol_seq
import scipy.stats

from bvbq import distributions,utils
from bvbq import BVBQMixMVN
```

```python
def loggaussian(x,mu,sigma):
    return -0.5*((x-mu)/sigma)**2 - np.log(np.sqrt(2*np.pi)*sigma)

def baselogprob(params):
    mu,sigma = params['mu'],params['sigma']
    term1 = np.log(scipy.stats.gamma(5.0).pdf(sigma))
    term2 = loggaussian(mu,0,sigma)
    return term1 + term2

def basesample(n):
    sigma = scipy.stats.gamma(5.0).rvs(size=(n,1))
    mu = np.random.randn(n,1)*sigma
    res = {'mu':mu,'sigma':sigma}
    return res
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
samples = basesample(100)
ax.scatter(samples['mu'],samples['sigma'])
ax.set_xlabel('mu')
ax.set_ylabel('sigma')
ax.set_ylim(bottom=0.0)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

sigma_plot,mu_plot = np.meshgrid(np.linspace(0,7,101)[1:],np.linspace(-15,15))
params = {'mu':mu_plot,'sigma':sigma_plot}
zplot = baselogprob(params)
ax.plot_surface(mu_plot,sigma_plot,np.exp(zplot))
```

```python
ndata = 50
sigma = scipy.stats.gamma(5).rvs(size=(ndata,1))
mu = np.random.randn(ndata,1)*sigma
```

```python
bvbq = BVBQMixMVN(['mu','sigma'],[1,1],[(None,None),(0.0,None)])
params_init = {'mu':mu,'sigma':sigma}
params_evals = baselogprob(params_init)
bvbq.initialize_data(params_init,params_evals,empirical_params=True,kind='smatern32')
bvbq.initialize_components()
```

```python
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.scatter(bvbq.eval_params['mu'][:ndata],bvbq.eval_params['sigma'][:ndata],color='blue')
ax2.scatter(bvbq.warped_eval_params[:ndata,0],bvbq.warped_eval_params[:ndata,1],color='blue')
```

```python
sigma_new = scipy.stats.gamma(5).rvs(size=(5,1))
mu_new = np.random.randn(5,1)*sigma_new
params_new = {'mu':mu_new,'sigma':sigma_new}
evals_new = baselogprob(params_new)
bvbq.insert_new_evaluations(params_new,evals_new)
```

```python
for i in range(20):
    params = bvbq.new_evaluation_point(unwarped=True, name='MMLT')
    evals = baselogprob(params)
    bvbq.insert_new_evaluations(params,evals)
```

```python
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.scatter(bvbq.eval_params['mu'][:ndata],bvbq.eval_params['sigma'][:ndata],color='blue')
ax1.scatter(bvbq.eval_params['mu'][ndata:],bvbq.eval_params['sigma'][ndata:],color='red')
ax2.scatter(bvbq.warped_eval_params[:ndata,0],bvbq.warped_eval_params[:ndata,1],color='blue')
ax2.scatter(bvbq.warped_eval_params[ndata:,0],bvbq.warped_eval_params[ndata:,1],color='red')
```
